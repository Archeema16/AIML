```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
import seaborn as sns
```

```python
boston_data  = load_boston()
#As it is a dictionary so check it through
#boston_data
print(boston_data.keys())
#print(boston_data.data.shape)
#print(boston_data.feature_names)
#print(boston_data.DESCR)
#df = pd.DataFrame(data=boston_data)
#df.head()
```

```python
#bos = pd.DataFrame(boston_data.data)
#bos.columns = boston_data.feature_names
#Or as both way are same
bos = pd.DataFrame(data=boston_data.data,columns = boston_data.feature_names)

print(bos.head())
```

```python
#Adding Label column in Dataframe
bos['Price'] = boston_data.target
bos
```

```python
bos.columns
```

```python
X_Data = bos[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
#X_Data
Y_Data = bos['Price']
#Y_Data
```

```python
X_Data.info()
X_Data.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_Data, Y_Data, test_size=0.33, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
Model = LinearRegression()
```

```python
#Trained My Model
Model.fit(X_train,y_train)
```

```python
print(Model.intercept_)
print(Model.coef_)
```

```python
Coeff_Frame = pd.DataFrame(data=Model.coef_,index=X_train.columns,columns=['Coeff'] )
Coeff_Frame
```

```python
Predictions = Model.predict(X_test)
```

```python
#To visualize if we had made a starigth line model
plt.scatter(y_test,Predictions)
plt.show()
```

```python
#Now lets calculate loss function or efficiency of our model 
from sklearn import metrics
#Mean Absolute Error
print(metrics.mean_absolute_error(y_test,Predictions))
#Mean Square Error
print(metrics.mean_squared_error(y_test,Predictions))
#Square Root Mean square Error
print(np.sqrt(metrics.mean_squared_error(y_test,Predictions)))  
```
